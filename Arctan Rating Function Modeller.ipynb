```python
## DON'T TOUCH THIS CELL!

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
```

```python
## DON'T TOUCH THIS CELL!
def arctan(x, a, b):
     y = a*np.arctan(x/b)
     return y
```

```python
### SET DESIRED CHIPS VS RP POINTS HERE
Chips = [0, 10, 100, 500, 800, 1200, 1500]
RP = [0, 1, 3, 8.5, 10, 11, 12]

```

```python
#DON'T TOUCH ANYTHING IN THIS CELL 
popt, pcov = curve_fit(arctan, Chips, RP)
print(" a = %f, b = %f" % (popt[0], popt[1]))

a = popt[0]
b = popt[1]
x = np.linspace(-3000, 3000, 2000)
```

```python
### SET THE LAST 2 PARAMENTERS AS YOU WISH    
y = arctan(x, a, b)
```

```python
plt.figure(figsize=(25,10))
fig, ax = plt.subplots(figsize=(25,10))
for i, c in enumerate(Chips):
    ax.annotate("("+str(Chips[i])+","+str(RP[i])+")", (Chips[i]+30, RP[i]-0.75), fontsize = 16)
    plt.plot(Chips[i], RP[i], marker='o', markersize=10, color="red")

plt.xticks(np.arange(-3000, 3000, step=200))
plt.yticks(np.arange(-15, 15, step=1))
plt.plot(x,y, linewidth=3.0, label='fit')
plt.ylabel('RP Won/Lost')
plt.xlabel('Chips Won/Lost')
plt.grid()
plt.show()
```
